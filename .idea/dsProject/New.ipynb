```python
import matplotlib.pyplot as plt;
```

```python
y = [1, 4, 9, 16, 25,36,49, 64]
x1 = [1, 16, 30, 42,55, 68, 77,88]
x2 = [1,6,12,18,28, 40, 52, 65]
```

```python
plt.plot(y, marker='o', linestyle = '--', label='Cows')
plt.plot(x1, marker='*', linestyle='-', label='Pigs')
plt.plot(x2, marker='^', linestyle=':', label='Chickens')

plt.xlabel('Animals')
plt.ylabel('Food_Distribution')
plt.title('Overall Animal Food intake sales')

plt.legend(loc='upper left')
plt.show()
```
